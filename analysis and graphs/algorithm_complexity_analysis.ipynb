# Algorithmic Complexity Analysis

In this analysis, we explored the algorithmic complexities of various key algorithms. We evaluated and compared linear O(n), constant (O(1)), and logarithmic (O(log n)) complexities to better understand the performance and efficiency of each algorithm in different scenarios.

## Results

Below are some key conclusions from the analysis:

### Linear Complexity (O(n))

- Algorithms whose complexity increases linearly with the input size were examined.
- We observed how the runtime grows proportionally to the size of the input data.

### Constant Complexity (O(1))

- Algorithms with constant complexity, meaning their runtime does not depend on the input size, were analyzed.
- These algorithms are highly efficient and offer constant performance regardless of data volume.

### Logarithmic Complexity (O(log n))

- Algorithms whose runtime grows logarithmically in relation to the input size were explored.
- The efficiency of these algorithms stands out, especially with larger datasets.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.arange(1, 11)  # Values of x (1 to 10)
y = x  # Linear function O(n) in this case

# Create the plot
plt.plot(x, y, label='y = x', marker='o')

# Configure the plot
plt.title('Linear Function O(n)')
plt.xlabel('n')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.arange(1, 11)  # Values of x (1 to 10)
y = np.ones_like(x)  # Array of ones with the same shape as x

# Create the plot in a Cartesian coordinate system
fig, ax = plt.subplots()

ax.plot(x, y, label='y = 1', marker='o')

# Configure the plot
ax.set_title('Constant (O(1)) Function')
ax.set_xlabel('n')
ax.set_ylabel('y')
ax.legend()
ax.grid(True)

# Show the plot
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.arange(1, 101)  # Values of x (1 to 100)
y = np.log2(x)  # Logarithmic function O(log(n))

# Create the plot
plt.plot(x, y, label='y = log2(x)', marker='o')

# Configure the plot
plt.title('Logarithmic Function O(log(n))')
plt.xlabel('n')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
