## Iteration Analysis for Finding Square Roots

In this analysis, we explore the number of iterations required to find the square root of different numbers using an approximation method. Below, a graph is presented illustrating the relationship between the number for which the square root is being calculated and the required number of iterations.



```python
import matplotlib.pyplot as plt

# Input data
numbers = list(range(2, 5))  # Square roots of 2 to 4
iterations = [41417820, 73202194, 99997500] 

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(numbers, iterations, marker='o', linestyle='-', color='b')
plt.title('Number of Iterations to Find the Square Root of Different Numbers')
plt.xlabel('Number')
plt.ylabel('Iterations')
plt.ticklabel_format(style='plain', axis='y')  # Disable scientific notation on the y-axis
plt.grid(True)
plt.show()

```
